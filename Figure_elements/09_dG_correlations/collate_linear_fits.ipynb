```python
from lmfit.model import load_modelresult
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasets = ['DMS', 'NMR', 'NNensemble', 'endpt', 'endpt2']

save_params = []
for i in datasets:
    for j in datasets:
        if i != j:
            print(f'Loading model result for {i} vs {j}')
            try:
                model = load_modelresult(f'/Users/edricchoi/LucksLab Dropbox/LucksLab/Lucks_Lab_Papers/In_Progress/88_EKC_Probing_model/Manuscript/Data_elements/09_dG_correlations/4U_{i}_v_{j}_fit.sav')
                
                save_params.append([i, j, model.params['slope'].value, model.params['intercept'].value, model.params['slope'].stderr, model.params['intercept'].stderr, model.rsquared])
                
                print(f'Successfully loaded model for {i} vs {j}')
            except FileNotFoundError:
                print(f'Model file for {i} vs {j} not found.')
            except Exception as e:
                print(f'An error occurred while loading the model for {i} vs {j}: {e}')

# Convert the results to a DataFrame
results_df = pd.DataFrame(save_params, columns=['Dataset1', 'Dataset2', 'Slope', 'Intercept', 'Slope_StdErr', 'Intercept_StdErr', 'R_squared'])
results_df.to_csv('correlation_results.csv', index=False)
```
